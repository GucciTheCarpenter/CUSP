```python
e = np.random.normal(0,1,1000)
Y1 = e[0]
Y1
```

##### I was unable to successfuly get "statsmodels.tsa.ar_model" to work so I created a quick loop for each section in Problem 1 to generate the necessary inputs.

##### a. The coefficient was close to ro=0.5 (0.4718 ) and the 95% confidence, 0.417 - 0.527, included 0.5 

```python
# Yt = (ro * Yt_1) + e

ro = 0.5
Yt = [e[0]]

x = [] # Y(t-1) values

for i in e:
    Yt_1index = len(Yt)
    
    Yt_1val = Yt[Yt_1index - 1]
    Yt_val = (ro * Yt_1val) + i
    
    Yt.append(Yt_val)
    x.append(Yt_1val)
    
plt.plot(e, label='e')
plt.plot(Yt, label='Yt = (0.5 * Y(t-1)) + e')
plt.legend(loc='best')

model = sm.OLS(Yt[1:], x)
results = model.fit()
print (results.summary())
```

# Problem 2 - Independent Random Walks

##### a. two independent random walks of 1,000 observations; bivariate linear model that relates Walk1 to Walk2

```python
Walk1 = np.random.normal(0,1,1000)
Walk1 = np.cumsum(Walk1)
Walk2 = np.random.normal(0,1,1000)
Walk2 = np.cumsum(Walk2)
```

```python
plt.plot(Walk1, label='Walk1')
plt.plot(Walk2, label='Walk2')
plt.legend(loc='best')

model = sm.OLS(Walk1, Walk2)
results = model.fit()
print (results.summary())
```

##### b. Monte Carlo simulation: 1,000 pairs of random walks; each time recording the estimated value of the slope coefficient of the bivariate regression; generate a histogram

```python
Monte_count = range(1000)
Monte_coef = []

for i in Monte_count:
    Walk1 = np.random.normal(0,1,1000)
    Walk1 = np.cumsum(Walk1)
    Walk2 = np.random.normal(0,1,1000)
    Walk2 = np.cumsum(Walk2)
    
    m, b = np.polyfit(Walk1, Walk2, 1)
    Monte_coef.append(m)

# print len(Monte_coef)
plt.hist(Monte_coef, bins=50)
```
