```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_NY = pd.read_csv('02_TechProfile_NY.csv')
df_NY.head()
```

```python
df_NY['Class %'] = ((df_NY['Total']) / float(df_NY['Total'][-1:])) * 100
df_NY['Class IDX of Top'] = ((df_NY['Class %']) / float(df_NY['Class %'][0]))
df_NY_top10 = df_NY[:10]
df_NY_top10
```

```python
n_groups = 10
index = np.arange(n_groups)
plt.bar(index, df_NY_top10['Total'])
plt.title('New York: Top 10 Total Patents, 2000 - 2011')
plt.ylabel('Patent Count')
plt.xlabel('Patent Category')
plt.xticks(index, (df_NY_top10['Class']), rotation=45)
grid('on')
plt.tight_layout()
```

```python
# plt.plot(df_NY_top10[['2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008','2009', '2010', '2011']], label = ['2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008','2009', '2010', '2011'])
index = np.arange(10)
plt.plot(df_NY_top10['2000'], label='2000')
plt.plot(df_NY_top10['2001'], label='2001')
plt.plot(df_NY_top10['2002'], label='2002')
plt.plot(df_NY_top10['2003'], label='2003')
plt.plot(df_NY_top10['2004'], label='2004')
plt.plot(df_NY_top10['2005'], label='2005')
plt.plot(df_NY_top10['2006'], label='2006')
plt.plot(df_NY_top10['2007'], label='2007')
plt.plot(df_NY_top10['2008'], label='2008')
plt.plot(df_NY_top10['2009'], label='2009')
plt.plot(df_NY_top10['2010'], label='2010')
plt.plot(df_NY_top10['2011'], label='2011')
plt.title('New York: Top 10 Patent Categories 2000 - 2011')
plt.ylabel('Patent Count')
plt.xlabel('Patent Category')
plt.xticks(index,(df_NY_top10['Class']), rotation=45)
grid('on')
legend()
```
