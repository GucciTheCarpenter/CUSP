```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
MSA_PatentsPer = pd.read_csv('01_PatentsPerMSA.csv')
MSA_PatentsPer.head()
```

```python
MSA_PatentsPer.columns = ['U.S. Regional Level', 'ID Code', 'ID MOD', 'U.S. Regional Title', 'Pat_2000', 'Pat_2001', 'Pat_2002', 'Pat_2003', 'Pat_2004', 'Pat_2005', 'Pat_2006', 'Pat_2007', 'Pat_2008', 'Pat_2009', 'Pat_2010', 'Pat_2011', 'Pat_Total']
MSA_PatentsPer.head()
```

```python
MSA_GDP = pd.read_csv('01_PerCapitalGDPbyMSA(2009 dollars)2.csv')
MSA_GDP.head()
```

```python
MSA_GDP.columns = ['Fips', 'Area', 'GDP_2001', 'GDP_2002', 'GDP_2003', 'GDP_2004', 'GDP_2005', 'GDP_2006', 'GDP_2007', 'GDP_2008', 'GDP_2009', 'GDP_2010', 'GDP_2011', 'GDP_2012', 'GDP_2013'
]
MSA_GDP.head()
```

```python
MSA_PatentsPer.shape, MSA_GDP.shape
```

```python
MSA_PatentsPer['ID MOD_str'] = MSA_PatentsPer['ID MOD'].astype(str)
```

```python
type(MSA_GDP['GDP_2001'][0]), type(MSA_PatentsPer['Pat_2000'][0])
```

```python
MSA_PatentsPer[['ID MOD', 'ID MOD_str']][:2]
```

```python
# data['result'] = data['result'].map(lambda x: x.lstrip('+-'))
MSA_PatentsPer['ID MOD_str'] = MSA_PatentsPer['ID MOD_str'].map(lambda x: x.rstrip('.'))
```

```python
MSA_GDP_Pat = pd.merge(MSA_GDP, MSA_PatentsPer, left_on='Fips', right_on='ID MOD_str')
MSA_GDP_Pat.head()
```

```python
MSA_GDP_Pat['Pat Avg 2001-2005'] = (MSA_GDP_Pat['Pat_2001'] + MSA_GDP_Pat['Pat_2002'] + MSA_GDP_Pat['Pat_2003'] + MSA_GDP_Pat['Pat_2004'] + MSA_GDP_Pat['Pat_2005'])/5.0
MSA_GDP_Pat[['Fips', 'Pat Avg 2001-2005']]
```

```python
MSA_GDP_Pat['GDP Avg 2008-2012'] = (MSA_GDP_Pat['GDP_2008'] + MSA_GDP_Pat['GDP_2009'] + MSA_GDP_Pat['GDP_2010'] + MSA_GDP_Pat['GDP_2011'] + MSA_GDP_Pat['GDP_2012'])/5.0
MSA_GDP_Pat[['Fips', 'Pat Avg 2001-2005', 'GDP Avg 2008-2012']]
```

```python
plt.scatter(MSA_GDP_Pat['Pat Avg 2001-2005'], MSA_GDP_Pat['GDP Avg 2008-2012'])
plt.title('MSA Patenting and Economic Performance')
plt.xlabel('Patents (2001 - 2005 avg)')
plt.xscale('log')
plt.ylabel('GDP per Capita (2008 - 2012 avg)')
# plt.yscale('log')
```
